Опишем трансформер, который будет подготавливать наши картинки к работе с нейросеточкой. Подготовка картинки состоит из трех пунктов:

```
IMG_SIZE = 128

transformer = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor()
])
```

Теперь опишем функции потерь . Согласно статье, для содержимого имеем следующее выражение:
$$L_{content}(\vec{p}, \vec{x}, l) = \sum_{i, j}{(F_{ij}^l - P_{ij}^l)}^2$$
где $\vec{p}$ и $\vec{x}$ - оригинальное и генерированное изображения, соответственно, а $P^l$ и $F^l$ - $l$-ый слои их тензоров признаков. Получаем:

```
class ContentLoss(nn.Module):
    def __init__(self, target,):
        super(ContentLoss, self).__init__()
        self.target = target.detach()  # Чтобы не получать ошибку от pytorch, убираем из дерева вычислений
        self.loss = F.mse_loss(self.target, self.target)  # Инициализируем лосс

    def forward(self, input_data):
        self.loss = F.mse_loss(input_data, self.target)
        return input_data
```

Теперь разберемся с функцией потерь для стиля изображения. Введем матрицу Грама, которая будет показывать корреляцию между стилевым и сгенерированным изображениеми: 
$$ G_{ij}^l = \sum_{k}{F_{ik}^l F_{jk}^l} $$
Тогда минимизируя MSE-расстояние между элементами матрицы Грама и оригинальным изображением, получим стиль, похожий на стиль стилевого изображения.
Пусть $\vec{a}$ и $\vec{x}$ - оригинальное и генерированное изображения, соответственно, а $A^l$ и $G^l$ - $l$-ый слои их тензоров признаков, тогда вклад этого слоя:
$$E_l = \frac{1}{4 N_l^2 M_l^2}\sum_{i, j}{(G_{ij}^l - A_{ij}^l)}^2$$
Тогда итоговый лосс:
$$L_{style}(\vec{a}, \vec{x}) = \sum_{l=0}^L{\omega_l E_l}$$
где $\omega_l$ - вес вклада каждого слоя в итоговый лосс. Получаем:

```
def gram_matrix(input_data):
    """
    batch_size - размер батча (в нашем случае всегда равен 1)
    h, w - размеры признаковой карты
    l - номер признаковой карты
    """
    batch_size, h, w, l = input_data.size()
    features = input_data.view(batch_size * h, w * l)  # транспонируем

    G = torch.mm(features, features.t()).div(4 * h * w)  # вычисляем матрицу + нормируем
    return G

class StyleLoss(nn.Module):
    def __init__(self, target_feature):
        super(StyleLoss, self).__init__()
        self.target = gram_matrix(target_feature).detach()
        self.loss = F.mse_loss(self.target, self.target)  # данные для инициализации

    def forward(self, input_data):
        G = gram_matrix(input_data)
        self.loss = F.mse_loss(G, self.target)
        return input_data
```

Задача данных слоя - посчитать лосс, поэтому для метода `forward` просто возвращаем полученное на вход значение, но при этом обновляем лосс слоя.


При тренировке модели VGG каждое изображение на котором она обучалась было нормировано по всем каналам (RGB), поэтому мы должны реализовать нормировку и для наших изображений тоже.

```
cnn_normalization_mean = torch.tensor([0.485, 0.456, 0.406]).to(device)
cnn_normalization_std = torch.tensor([0.229, 0.224, 0.225]).to(device)

class Normalization(nn.Module):
    def __init__(self, mean, std):
        super(Normalization, self).__init__()
        # Используем .view чтобы напрямую работать с тензором изображения
        self.mean = torch.tensor(mean).view(-1, 1, 1)
        self.std = torch.tensor(std).view(-1, 1, 1)

    def forward(self, img):
        return (img - self.mean) / self.std
```

Определим после каких уровней мы будем счиатать ошибки стиля, а после каких ошибки контента:

```
content_layers = ['conv_4']
style_layers = ['conv_1', 'conv_2', 'conv_3', 'conv_4', 'conv_5']
```

Загрузим предобученную модель VGG19:

```
cnn = models.vgg19(pretrained=True).features.to(device).eval()
```

Теперь соберем это все в класс, который отдаст на выходе модель и будет хранить значения двух функции потерь.

```
class StyleTransfer:
    def __init__(self, cnn, normalization_mean, normalization_std, content_img, style_img, input_img, num_steps=500, style_weight=100000, content_weight=1):
        print('Building the style transfer model ...')
        model, style_losses, content_losses = self.get_style_model_and_losses(cnn, normalization_mean, normalization_std, style_img, content_img)
        optimizer = self.get_input_optimizer(input_img)

        print('Optimizing ...')
        pbar = tqdm(total=num_steps+1)
        curr_step = [0]
        while curr_step[0] <= num_steps:
            def closure():
                # correct the values 
                # это для того, чтобы значения тензора картинки не выходили за пределы [0;1]
                input_img.data.clamp_(0, 1)

                optimizer.zero_grad()

                model(input_img)

                style_score = 0
                content_score = 0

                for sl in style_losses:
                    style_score += sl.loss
                for cl in content_losses:
                    content_score += cl.loss
                
                #взвешивание ощибки
                style_score *= style_weight
                content_score *= content_weight

                loss = style_score + content_score
                loss.backward()

                curr_step[0] += 1
                pbar.update(1)
                if curr_step[0] % 50 == 0:
                    print("Step {}:".format(curr_step[0]))
                    print('Style Loss : {:4f} Content Loss: {:4f}'.format(
                        style_score.item(), content_score.item()))
                    print()

                return loss

            optimizer.step(closure)

        # a last correction...
        input_img.data.clamp_(0, 1)

        self.result =  input_img
    
    def get_input_optimizer(self, input_img):
        # this line to show that input is a parameter that requires a gradient
        #добоваляет содержимое тензора катринки в список изменяемых оптимизатором параметров
        optimizer = optim.LBFGS([input_img.requires_grad_()]) 
        return optimizer

    def get_style_model_and_losses(self, cnn, normalization_mean, normalization_std, style_img, content_img,
                                   content_layers=['conv_4'], 
                                   style_layers=['conv_1', 'conv_2', 'conv_3', 'conv_4', 'conv_5']):
            cnn = copy.deepcopy(cnn)

            # normalization module
            normalization = Normalization(normalization_mean, normalization_std).to(device)

            # just in order to have an iterable access to or list of content/syle
            # losses
            content_losses = []
            style_losses = []

            # assuming that cnn is a nn.Sequential, so we make a new nn.Sequential
            # to put in modules that are supposed to be activated sequentially
            model = nn.Sequential(normalization)

            i = 0  # increment every time we see a conv
            for layer in cnn.children():
                if isinstance(layer, nn.Conv2d):
                    i += 1
                    name = 'conv_{}'.format(i)
                elif isinstance(layer, nn.ReLU):
                    name = 'relu_{}'.format(i)
                    # The in-place version doesn't play very nicely with the ContentLoss
                    # and StyleLoss we insert below. So we replace with out-of-place
                    # ones here.
                    #Переопределим relu уровень
                    layer = nn.ReLU(inplace=False)
                elif isinstance(layer, nn.MaxPool2d):
                    name = 'pool_{}'.format(i)
                elif isinstance(layer, nn.BatchNorm2d):
                    name = 'bn_{}'.format(i)
                else:
                    raise RuntimeError('Unrecognized layer: {}'.format(layer.__class__.__name__))

                model.add_module(name, layer)

                if name in content_layers:
                    # add content loss:
                    target = model(content_img).detach()
                    content_loss = ContentLoss(target)
                    model.add_module("content_loss_{}".format(i), content_loss)
                    content_losses.append(content_loss)

                if name in style_layers:
                    # add style loss:
                    target_feature = model(style_img).detach()
                    style_loss = StyleLoss(target_feature)
                    model.add_module("style_loss_{}".format(i), style_loss)
                    style_losses.append(style_loss)

            # now we trim off the layers after the last content and style losses
            #выбрасываем все уровни после последенего styel loss или content loss
            for i in range(len(model) - 1, -1, -1):
                if isinstance(model[i], ContentLoss) or isinstance(model[i], StyleLoss):
                    break

            model = model[:(i + 1)]

            return model, style_losses, content_losses
```

```
input_img = content_img.clone()
output = StyleTransfer(cnn, cnn_normalization_mean, cnn_normalization_std, content_img, style_img, input_img).result
```

```
plt.figure()
img_show(content_img, title='Input Image')
img_show(input_img, title='Result Image')
```

Теперь модернизируем класс для работы с двумя стилями. Для этого нам понадибится еще одно слагаемое в лоссе

```
style1_img = image_loader("style_images/flat.jpg")
style2_img = image_loader("style_images/face.jpg")
content_img = image_loader("content_images/eifel.jpg")
```

```
class StyleTransfer2:
    def __init__(self, cnn, normalization_mean, normalization_std, content_img, style1_img, style2_img, input_img, num_steps=500, style1_weight=100000, style2_weight=100000, content_weight=1):
        print('Building the style transfer model ...')
        model, style1_losses, style2_losses, content_losses = self.get_style_model_and_losses(cnn, normalization_mean, normalization_std, style1_img, style2_img, content_img)
        optimizer = self.get_input_optimizer(input_img)

        print('Optimizing ...')
        pbar = tqdm(total=num_steps+1)
        curr_step = [0]
        while curr_step[0] <= num_steps:
            def closure():
                # correct the values 
                # это для того, чтобы значения тензора картинки не выходили за пределы [0;1]
                input_img.data.clamp_(0, 1)

                optimizer.zero_grad()

                model(input_img)

                style1_score = 0
                style2_score = 0
                content_score = 0

                for sl in style1_losses:
                    style1_score += sl.loss
                for sl in style2_losses:
                    style2_score += sl.loss
                for cl in content_losses:
                    content_score += cl.loss
                
                #взвешивание ощибки
                style1_score *= style1_weight
                style2_score *= style2_weight
                content_score *= content_weight

                loss = style1_score + style2_score + content_score
                loss.backward()

                curr_step[0] += 1
                pbar.update(1)
                if curr_step[0] % 50 == 0:
                    print("Step {}:".format(curr_step[0]))
                    print('Style1 Loss : {:4f} Style2 Loss : {:4f}  Content Loss: {:4f}'.format(
                        style1_score.item(), style2_score.item(), content_score.item()))
                    print()

                return loss

            optimizer.step(closure)

        # a last correction...
        input_img.data.clamp_(0, 1)

        self.result =  input_img
    
    def get_input_optimizer(self, input_img):
        # this line to show that input is a parameter that requires a gradient
        #добоваляет содержимое тензора катринки в список изменяемых оптимизатором параметров
        optimizer = optim.LBFGS([input_img.requires_grad_()]) 
        return optimizer

    def get_style_model_and_losses(self, cnn, normalization_mean, normalization_std, style1_img, style2_img, content_img,
                                   content_layers=['conv_4'], 
                                   style_layers=['conv_1', 'conv_2', 'conv_3', 'conv_4', 'conv_5']):
            cnn = copy.deepcopy(cnn)

            # normalization module
            normalization = Normalization(normalization_mean, normalization_std).to(device)

            # just in order to have an iterable access to or list of content/syle
            # losses
            content_losses = []
            style1_losses = []
            style2_losses = []

            # assuming that cnn is a nn.Sequential, so we make a new nn.Sequential
            # to put in modules that are supposed to be activated sequentially
            model = nn.Sequential(normalization)

            i = 0  # increment every time we see a conv
            for layer in cnn.children():
                if isinstance(layer, nn.Conv2d):
                    i += 1
                    name = 'conv_{}'.format(i)
                elif isinstance(layer, nn.ReLU):
                    name = 'relu_{}'.format(i)
                    # The in-place version doesn't play very nicely with the ContentLoss
                    # and StyleLoss we insert below. So we replace with out-of-place
                    # ones here.
                    #Переопределим relu уровень
                    layer = nn.ReLU(inplace=False)
                elif isinstance(layer, nn.MaxPool2d):
                    name = 'pool_{}'.format(i)
                elif isinstance(layer, nn.BatchNorm2d):
                    name = 'bn_{}'.format(i)
                else:
                    raise RuntimeError('Unrecognized layer: {}'.format(layer.__class__.__name__))

                model.add_module(name, layer)

                if name in content_layers:
                    # add content loss:
                    target = model(content_img).detach()
                    content_loss = ContentLoss(target)
                    model.add_module("content_loss_{}".format(i), content_loss)
                    content_losses.append(content_loss)

                if name in style_layers:
                    # add style loss:
                    target_feature1 = model(style1_img).detach()
                    target_feature2 = model(style2_img).detach()
                    style1_loss = StyleLoss(target_feature1)
                    style2_loss = StyleLoss(target_feature2)
                    model.add_module("style1_loss_{}".format(i), style1_loss)
                    model.add_module("style2_loss_{}".format(i), style2_loss)
                    style1_losses.append(style1_loss)
                    style2_losses.append(style2_loss)

            # now we trim off the layers after the last content and style losses
            #выбрасываем все уровни после последенего styel loss или content loss
            for i in range(len(model) - 1, -1, -1):
                if isinstance(model[i], ContentLoss) or isinstance(model[i], StyleLoss):
                    break

            model = model[:(i + 1)]

            return model, style1_losses, style2_losses, content_losses
```
